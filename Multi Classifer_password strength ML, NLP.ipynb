```python
# instead of tokeniazing the whole passwords i.e over 6 lakhs wods, just taking the char from a-z
def word_divide_char(inputs):
    characters=[]
    for i in inputs:
        characters.append(i)
    return characters
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(tokenizer=word_divide_char) #tokenizer are the charecters
X=vectorizer.fit_transform(X)
```

```python
print(vectorizer.vocabulary_, end='')
```

```python
X.shape # before creating the vectors we had only 2 features 
```

```python
data.iloc[0,0]
```

```python
feature_names = vectorizer.get_feature_names()
 
#get tfidf vector for first document
first_document_vector=X[0]
 
#print the scores
df = pd.DataFrame(first_document_vector.T.todense(), index=feature_names, columns=["tfidf"])
df.sort_values(by=["tfidf"],ascending=False).head(10) # rest all are zeros
```

```python
## Logistics Regression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)  #splitting
# X is vector of chars
```

```python
#Logistic reg is basic is for binary classfn, parameters used are for multi_class.
#ovr is one verus rest OR multi nominal. penalty can be either l1 or l2 for penalization
log_class=LogisticRegression(penalty='l2',multi_class='ovr') # press shift n tab
log_class.fit(X_train,y_train)
```

In the multiclass case, the training algorithm uses the one-vs-rest (OvR)
scheme if the 'multi_class' option is set to 'ovr', and uses the
cross-entropy loss if the 'multi_class' option is set to 'multinomial'.

```python
print('multi_class with OvR: ',log_class.score(X_test,y_test))
```

```python
## trying multinomial
## takes long time
clf = LogisticRegression(random_state=0, multi_class='multinomial', solver='newton-cg')
clf.fit(X_train, y_train) #training
print(clf.score(X_test, y_test))
```

```python
X_predict=np.array(["!@#$%^&*()1d"])
X_predict=vectorizer.transform(X_predict)
y_pred=log_class.predict(X_predict)
print(y_pred)
```

```python
# 0='week', 1='medium', 2='strong'
```

```python
# Training model using Naive bayes classifier
from sklearn.naive_bayes import MultinomialNB
password_strength_model = MultinomialNB().fit(X_train, y_train)
y_pred=password_strength_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
#from sklearn.metrics import roc_auc_score
from sklearn import metrics
#print('password strength test set roc-auc:', metrics.roc_auc_score(y_test, y_pred)) # did not support multi class
print('classification_report :- ')
print(classification_report(y_test, y_pred))
```

```python
# Plot Confusion Matrix
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test, y_pred)
#print('confusion_matrix',mat)
names = np.unique(y_pred)
#Heat map works by correlation. This shows you which variables are correlated to each other 
#from a scale of 1 being the most correlated and -1 is not correlated at all
sns.heatmap(mat, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=names, yticklabels=names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
# calculate accuracy
from sklearn import metrics
print(metrics.accuracy_score(y_test, y_pred))
```
